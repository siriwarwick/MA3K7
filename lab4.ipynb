```python
# V1 Simple NIM with no strategy or checks
import numpy as np

rng = np.random.default_rng()

sticks = 10

print('|'*sticks, f'{sticks} left')

def game(myturn):
    global sticks
    if myturn == 1:
        x = int(input('Your turn: How many to remove (up to 3)?'))
        sticks = sticks - x
        print('|'*sticks, f'{sticks} left')
        if sticks==1:
            print('You win!')
            return 0
        game(0)
    
    else:
        x = rng.integers(1, min(sticks,4))
        print(f"Bot's turn: {x} removed")
        sticks = sticks - x
        print('|'*sticks, f'{sticks} left')
        if sticks==1:
            print('You lose!')
            return 0
        game(1)
    
myturn = 1 # 1 = player starts first, 0 = bot starts first

game(myturn)
```

```python
# V2 Simple NIM with input check (but no strategy)

import numpy as np

rng = np.random.default_rng()

sticks = 10
print('|'*sticks, f'{sticks} left')

def check(x, sticks):
    while True:
        if not (1<=x<=min(sticks,3)): 
            print("Nope... choose again")
            x = int(input('How many to remove (up to 3)?'))
        else: return x
    
def game(myturn):
    global sticks
    if myturn == 1:
        x = int(input('Your turn: How many to remove (up to 3)?'))
        x = check(x, sticks)
        sticks = sticks - x
        print('|'*sticks, f'{sticks} left')
        if sticks==1:
            print('You win!')
            return 0
        game(0)
    
    else:
        x = rng.integers(1, min(sticks,4))
        print(f"Bot's turn: {x} removed")
        sticks = sticks - x
        print('|'*sticks, f'{sticks} left')
        if sticks==1:
            print('You lose!')
            return 0
        game(1)
    
myturn = 1 # 1 = player starts first, 0 = bot starts first

game(myturn)
```

```python
# V3 Simple NIM with input check AND strategy (bot starts first and always wins
# unless starting pile is a multiple of 4)
# Strategy = always leaves pile with 4n+1 sticks until the last move
# See condensed version by Awesome GameDev https://www.youtube.com/watch?v=vnYES_fba-M

import numpy as np

rng = np.random.default_rng()

sticks = 10 
print('|'*sticks, f'{sticks} left')

def check(x, sticks):
    while True:
        if not (1<=x<=min(sticks-1,3)): 
            print("Nope... choose again")
            x = int(input('How many to remove (up to 3)?'))
        else: return x
    
def strategy(sticks): 
    N = sticks%4 
    if N==1: 
        x = rng.integers(1, min(sticks,4)) # too bad.. no advantage, so random move
    else: 
        x= (N-1)%4
    return x
   
def game(myturn):
    global sticks
    if myturn == 1:
        x = int(input('Your turn: How many to remove (up to 3)?'))
        x = check(x, sticks)
        sticks = sticks - x
        print('|'*sticks, f'{sticks} left')
        if sticks==1:
            print('You win!')
            return 0
        game(0)
    
    else:
        x = strategy(sticks)
        print(f"Bot's turn: {x} removed")
        sticks = sticks - x
        print('|'*sticks, f'{sticks} left')
        if sticks==1:
            print('You lose!')
            return 0
        game(1)
    
myturn = 0    # 0 = bot starts first and always wins

game(myturn)
```
